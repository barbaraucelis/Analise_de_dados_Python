```python
import pandas as pd
```

```python
# Importar colunas específicas do arquivo: 2 - Estabelecimento_CNES; 3 - Complexidade, Carater Atendimento, 7 - Sub Grupo - Procedimento, 9 - Procedimento

df = pd.read_csv('sia112018.csv', sep = ';', encoding = 'cp1252', usecols = [2, 3, 12, 7, 9])
```

```python
df.head()
```

```python
#Trocar nome das colunas
df.columns = ['Estabelecimento', 'Complexidade', 'Subgrupo Procedimento', 'Procedimento', 'Carater']
```

```python
df.head()
```

```python
#Descrever as colunas numéricas
df.describe()
```

```python
#Listas os estabelcimentos presentes nos dados
df['Estabelecimento'].unique()
```

```python
# Quantos procedimentos foram realizados em cada estabelecimento?
df['Estabelecimento'].value_counts()
```

```python
# Quantos procedimentos por SubGrupo Procedimento
df['Subgrupo Procedimento'].value_counts()
```

```python
# Plotar o gráfico por SubGrupo Procdimento
df['Subgrupo Procedimento'].value_counts().plot.pie()
```

```python
# Vamos criar um subconjunto dos dados originais
df_HMIB = df[df['Estabelecimento'] == 'HOSPITAL MATERNO INFANTIL DE BRASILIA - HMIB']
```

```python
df_HMIB.head(2)
```

```python
# Listar Linhas aleatórias do DataFrame
df_HMIB.sample(5)
```

```python
# A quantidade de procdimentos realizados no HMIB
df_HMIB['Procedimento'].value_counts()
```

```python
# Subdividr o dataframe com base em pedaço de palavra
df_HMIB[df_HMIB['Procedimento'].str.contains('Exame')].count()
```

```python
# Dividir o Dataframe original apenas pelo procedimento de 'Consulta Medica em Atenção Especializada'
df_CMAE = df[df['Procedimento'] == 'Consulta Medica em Atenção Especializada']
```

```python
df_CMAE.head(2)
```

```python
# Verificar a quantidade de Consulta Medica em Atenção Especializada por estabelecimento
df_CMAE['Estabelecimento'].value_counts()
```

```python
# Verificar a quantidade de Consulta Medica em Atenção Especializada por estabelecimento por caráter de atendimento

df_CMAE['Carater'].value_counts()
```

```python
# Plotar um gráfico de barras horizontal
df_CMAE['Estabelecimento'].value_counts().plot.barh()
```

```python
# Melhorar o gráfico de barras horizontal, com ordenação e titulo
df_CMAE['Estabelecimento'].value_counts(ascending=True).plot.barh(title='Quantidade de Consultas Especializadas por Estabelecimento')
```
