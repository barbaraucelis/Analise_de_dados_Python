# Análise dos dados da API de partidos
## Análise de dados obtidos
Link: https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla

```python
import pandas as pd
import requests
```

```python
# Obter os dados a partir de uma API
# Variável para URL
url = 'https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla'
```

```python
# Obtendo os dados 
dados = requests.get(url)
```

```python
# Verificando a resposta
dados.text
```

```python
# converter os dados da API para DataFrame
df = pd.DataFrame(dados.json()['dados'])
```

```python
df
```

```python
# Qual o número atual de partidos?
df['sigla'].count()
```

```python
df['sigla'].dtype()
```

```python
# Qualquer campo do dataframe pode ser obtido usando as funções de localização
# O iloc retorna um campo com base no índice [linha, coluna]
df.iloc[0,2]
```

```python
# O loc retorna o campo com base no nome da coluna
df.loc[0, 'uri']
```
