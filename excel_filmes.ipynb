```python
import pandas as pd
%matplotlib inline

#Criar Dataframe com base em arquivo excel e pular as 2 primeiras linhas (skiprows) e as 20 últimas linhas (skipfooter)
df = pd.read_excel('listagem-de-filmes-brasileiros-e-estrangeiros-exibidos-2009-a-2019.xlsx', skiprows=2, skipfooter=20)
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
# Converter tipos de dados
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Identificar o problema da coluna Data de Lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que estão com relançamento
df[df['Data de lançamento'] == 'Relançamento'].head()
```

## Decisão sobre dados incorretos/faltantes/divergentes

```python
# Criar um novo DataFrame sem os Relançamentos
df_novo = df[(df['Data de lançamento'] != 'Relançamento') & (df['Data de lançamento'] != 'relançamento') & (df['Renda (R$) no ano de exibição'] != 'ND') & (df['Público no ano de exibição'] != 'ND')].copy()
```

```python
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
df_novo['Público no ano de exibição'] = pd.to_numeric(df_novo['Público no ano de exibição'])
df_novo['Renda (R$) no ano de exibição'] = pd.to_numeric(df_novo['Renda (R$) no ano de exibição'])
```

```python
df_novo.dtypes
```

```python
df_novo['Renda (R$) no ano de exibição'].value_counts()
```

```python
# Quais os anos tiveram mais filmes lançados
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Plotar um gráfico com os lançamentos por ano
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
# Qual filme com maior bilheteria
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# Os 3 filmes com maior bilheteria
df_novo.nlargest(3, 'Renda (R$) no ano de exibição')
```

```python
# Qual o filme com menor bilheteria
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```
